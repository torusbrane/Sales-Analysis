## By manually combing through the rating column, we found that at index location 1279, the foreign character "|" was found. Through this manual process, we can drop the entire row to keep the data clean. We then reset the index because we essentially took a row out of the "stack". There are also 2 other options here. We can programatically remove the row that isn't a float (instead of doing it manually), or we can fill the missing data with the average rating (also programatically). We will do both, and discuss the caveats of using both.

```python
print(rating.loc[[1279]], "\n")
print("----------------------------\n")
rating_numerical = amazonSales.rating.drop([1279]).reset_index(drop=True)
print(rating_numerical)
```
