### Here, we will make a function that will change the column of choice (column_name) of the Dataframe (df) to all numeric values. All of the rows of the column that are non-numeric will be converted into NaN (Not a Number). Then, to convert it to a float, we do that with the `.astype()`. Lastly, we reset the index to ensure that it deletes the entire row to keep the data clean.

### First, we need to make a copy of the Dataframe so that any changes that we make, don't affect the original copy.

```python
amazonSales_1 = amazonSales.copy()
```

```python
def convert_column_to_float(df, column_name):
    """
    Convert a column in a DataFrame to numeric values, handling non-numeric values by 
    converting them to NaN, and then change the data type to float.

    Parameters:
        df: pandas.DataFrame
            the DataFrame to be used.
        
        column_name:  str
            the column to be converted to float.

    Returns:
        pandas.DataFrame: The DataFrame with the specified column converted to float.
    """
    # Convert column to numeric, setting errors='coerce' to handle non-numeric values by converting them to NaN
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
    
    # Drop rows where the specified column is NaN
    df.dropna(subset=[column_name], inplace=True)

    # Convert the column to float (optional, as it's likely already in float format after to_numeric)
    df[column_name] = df[column_name].astype(float)
    
    # Reset the index and drop the old index to avoid adding it as a column
    df.reset_index(drop=True, inplace=True)

    return df
```

```python
convert_column_to_float(amazonSales_1, "rating")
```

### Now we will do the same process, only we will fill the NaN with the mean (average) of the entire column. The last step that is also added is the rounding step. This ensures that only one digit will be saved and displayed after the averaging process.

### Like last time, we will make a copy of the Dataframe.

```python
amazonSales_2 = amazonSales.copy()
```

```python
def convert_column_to_float_fillna_with_mean(df, column_name):
    """
    Convert a column in a DataFrame to numeric values, converting non-numeric values to NaN, 
    and then change the data type to float. NaN values are then replaced with 
    the average of the column.

    Parameters:
        df: pandas.DataFrame
            The DataFrame to be used.
        column_name: str
            The name of the column to convert.

    Returns:
        pandas.DataFrame
            The DataFrame with the specified column converted to float and NaN filled with the 
            column's average.
    """
    # Convert column to numeric, setting errors='coerce' to handle non-numeric values by converting them to NaN
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce')

    # Calculate the mean of the column, excluding NaN values
    mean_value = df[column_name].mean()

    # Fill NaN values in the column with the calculated mean
    df[column_name].fillna(value=mean_value, inplace=True)
    
    # Round the column values to 1 decimal place
    df[column_name] = df[column_name].round(1)

    return df
```

```python
convert_column_to_float_fillna_with_mean(amazonSales_2, "rating")
```

```python
amazonSales_2.rating[1277:1283]
```

### So the pros and cons:
---
* The pro of deleting the contaminated rows - keeps the data clean with no bias (mean, median or mode) injected into the dataset
* The con of deleting the contaminated rows - if there is not enough data, it can give less accurate results by deleting the rows.

---
* The pro of filling with average - helps to fill in the missing data to keep deleted data to a minimum, which allows for closer approximation of the missing data. This can help when modelling
the data, which can help spot things like outliers.
* The con of filling with average - may add some accidental bias into the dataset that was not there in the beginning.
---

### There are also other ways to fill in the missing data, like using 'SimpleImputer' provided by scikit-learn, but we will use that for another dataset when modeling. For now, we will just use what we have now since we're just analyzing the data.

### It looks like 'rating_count' is comprised of just integers, so let's convert that column to integers. Keeping in mind that we need to be mindful of the commas in the larger numbers. If not, we may accidentally erase numbers larger than 999.

```python
def convert_column_to_int(df, column_name):
    '''
    Convert the column from the dataframe to pure numeric data, any non-numeric data will be converted to
    NaN, the NaN's will be deleted, and the remaining numeric data will be converted to integers. All
    commas will be deleted to prevent any deletion of numbers higher than 999. (Ex: 1,000)
    
    Parameters:
        df: pandas.DataFrame
            The DataFrame to be used.

        column_name: str
            The column to be converted to int.
            
    Returns:
        pandas.DataFrame
            The DataFrame with the specified column converted to int.
    '''
    # Convert the column to string, remove the commas, convert column to numeric, 
    # setting errors='coerce' to handle non-numeric values by converting them to NaN
    df[column_name] = df[column_name].astype(str).str.replace(',', '').apply(pd.to_numeric, errors='coerce')
        
    # Drop rows where the specified column is NaN
    df.dropna(subset=[column_name], inplace=True)

    # Convert the column to int
    df[column_name] = df[column_name].astype(int)
    
    # Reset the index and drop the old index to avoid adding it as a column
    df.reset_index(drop=True, inplace=True)

    return df
```

### Now we will also remove the currency symbol, along with the comma(s) from the  'discounted_price' and 'actual_price' columns.

```python
def convert_column_to_int_remove_currency_symbols(df, column_name):
    '''
    Convert the column from the dataframe to pure numeric data, any non-numeric data will be converted to
    NaN, the NaN's will be deleted, and the remaining numeric data will be converted to integers. All
    commas and leading currency symbols will be deleted.
    
    Parameters:
        df: pandas.DataFrame
            The DataFrame to be used.

        column_name: str
            The column to be converted to int.
            
    Returns:
        pandas.DataFrame
            The DataFrame with the specified column converted to int.
    '''
    # Convert the column to string, remove the currency symbols and commas, convert column to numeric, 
    # setting errors='coerce' to handle non-numeric values by converting them to NaN
    df[column_name] = df[column_name].astype(str).str.replace(r'^\D+', '', regex=True).str.replace(',', '')
    
    df[column_name]= pd.to_numeric(df[column_name], errors='coerce')
        
    # Drop rows where the specified column is NaN
    df.dropna(subset=[column_name], inplace=True)

    # Convert the column to int
    df[column_name] = df[column_name].astype(int)
    
    # Reset the index and drop the old index to avoid adding it as a column
    df.reset_index(drop=True, inplace=True)

    return df
```

```python
convert_column_to_int_remove_currency_symbols(amazonSales_1, 'discounted_price')
```

```python
convert_column_to_int_remove_currency_symbols(amazonSales_1, 'actual_price')
```

### Now let's convert the 'discount_percentage' column

```python
def convert_column_to_int_remove_percent(df, column_name):
    '''
    Convert the column from the dataframe to pure numeric data, any non-numeric data will be converted to
    NaN, the NaN's will be deleted, and the remaining numeric data will be converted to integers. All
    commas and trailing symbols will be deleted.
    
    Parameters:
        df: pandas.DataFrame
            The DataFrame to be used.

        column_name: str
            The column to be converted to int.
            
    Returns:
        pandas.DataFrame
            The DataFrame with the specified column converted to int.
    '''
    # Convert the column to string, remove the percent sign and commas, convert column to numeric, 
    # setting errors='coerce' to handle non-numeric values by converting them to NaN
    df[column_name] = df[column_name].astype(str).str.replace(r'\D+$', '', regex=True).str.replace(',', '')
    
    df[column_name]= pd.to_numeric(df[column_name], errors='coerce')
        
    # Drop rows where the specified column is NaN
    df.dropna(subset=[column_name], inplace=True)

    # Convert the column to int
    df[column_name] = df[column_name].astype(int)
    
    # Reset the index and drop the old index to avoid adding it as a column
    df.reset_index(drop=True, inplace=True)

    return df
```
